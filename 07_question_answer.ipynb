## Tokenizing text for QA

````python
from transformers import AutoTokenizer

model_ckpt = "deepset/minilm-uncased-squad2"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
````

````python
question = "How much music can this hold?"

context = "An MP3 is about 1 MB/minute, so about 6000 hours depending on file size."

inputs = tokenizer(question, context, return_tensors="pt")
````

````python
from pprint import pprint
# a 0 indicates a question token, a 1 indicates a context token in token_types_ids
pprint(inputs)
````

````python
print(tokenizer.convert_ids_to_tokens(inputs.input_ids[0]))
````

````python
# Modeling
import torch
from transformers import AutoModelForQuestionAnswering

model = AutoModelForQuestionAnswering.from_pretrained(model_ckpt)

with torch.no_grad():
    outputs = model(**inputs)
print(outputs)
````

````python
start_logits = outputs.start_logits
end_logits = outputs.end_logits
````

````python
print(f"Input IDs shape: {inputs.input_ids.size()}")
print(f"Start logits shape: {start_logits.size()}")
print(f"End logits shape: {end_logits.size()}")
````

You're right to seek clarification on this point. Let me explain how documents are typically structured in open-domain and closed-domain QA systems, and contrast that with the SubjQA example:

1. Open-domain QA:
   - Documents are not typically associated with specific products.
   - Each document can cover a wide range of topics or information.
   - Documents might be encyclopedia articles, news articles, web pages, or any general text.
   - There's no inherent product-based structure to the document collection.

2. Closed-domain QA:
   - Documents are focused on a specific domain or topic, but still not usually associated with individual products.
   - For example, in a medical QA system, documents might be medical textbooks, research papers, or clinical guidelines.
   - The domain is restricted, but documents aren't tied to specific products.

3. SubjQA-like datasets:
   - Here, each document is associated with a single product.
   - This is a more specialized structure, often used in e-commerce or product review scenarios.
   - For example, a document might be a collection of reviews or specifications for a particular product.

The key difference is in the organization and purpose of the documents:

- In open and most closed-domain systems, documents are independent units of information.
- In product-specific datasets like SubjQA, documents have an additional layer of organization, being tied to specific products.

This product-specific structure requires special handling during evaluation, as you might want to test if the retriever can find the correct document not just based on relevance to the query, but also based on its association with the correct product. This is why the standard eval() method might not be suitable for such datasets, as it doesn't account for this product-based filtering.

````python
def evaluate_retriever(retriever, topk_values = [1,3,5,10, 20]):
    topk_result = {}
    max_topk = max(topk_values)
    pipe = DocumentSearchPipeline(retriever=retriever)
    eval_result = pipe.eval(
        labels=labels_agg,
        params={"Retriever": {"top_k": max_topk}}
    )

    for k in topk_values:
        metric = eval_result.calculate_metrics(simulated_top_k_retriever=k)
        topk_result[k] = {"recall": metric['Retriever']["recall_single_hit"]}
    return pd.DataFrame.from_dict(topk_result, orient="index")
````

````python
bm25_topk_df = evaluate_retriever(bm25_retriever)
````

````python
print(bm25_topk_df)
````

````python
def plot_retriever_eval(dfs, retriever_names):
    fig, ax = plt.subplots()
    for df, retriever_name in zip(dfs, retriever_names):
        df.plot(y="recall", ax=ax, label=retriever_name)
    plt.xticks(df.index)
    plt.ylabel("Top-k Recall")
    plt.xlabel("k")
    plt.show()

plot_retriever_eval([bm25_topk_df], ["BM25"])
````

to evaluate retriver we use `DocumentSearchPipeline` since it has only one node ie retriver


but there is nothing like DocumentSearchPipeline for reader,


so we use pipeline consist one node


since there is no retriver in the pipeline it will take revalent document directly at input

````python
from haystack.pipelines import Pipeline

def evaluate_reader(reader):
    score_keys = ["exact_match", "f1"]
    pipe = Pipeline()
    pipe.add_node(component=reader, name="Reader", inputs=["Query"])

    eval_result = pipe.eval(
        labels=labels_agg,
        documents=[[label.document for label in multilabel.labels] for multilabel in labels_agg],
        params={}
    )

    metrics = eval_result.calculate_metrics(simulated_top_k_reader=1)

    return {k:v for k,v in metrics["Reader"].items() if k in score_keys}
````

````python
def plot_reader_eval(reader_eval):
    fig, ax = plt.subplots()
    df = pd.DataFrame.from_dict(reader_eval).reindex(["exact_match", "f1"])
    df.plot(kind="bar", ylabel="Score", rot=0, ax=ax)
    ax.set_xticklabels(["EM", "F1"])
    plt.legend(loc='upper left')
    plt.show()

plot_reader_eval(reader_eval)
````

````python
def create_paragraphs(df):
    paragraphs = []
    id2context = dict(zip(df['review_id'], df['context']))
    for review_id, review in id2context.items():
        qas = []
        # filter for all question-answer pairs about a specific context
        review_df = df.query(f"review_id == '{review_id}'")
        id2question = dict(zip(review_df['id'], review_df['question']))
        # build up the qas array
        for qid, question in id2question.items():
            # Filter for single question ID
            question_df = df.query(f"id == '{qid}'").to_dict(orient="list")
            ans_start_idxs = question_df['answers.answer_start'][0].tolist()
            ans_text = question_df["answers.text"][0].tolist()
            # Fill the answerable questions
            if len(ans_start_idxs):
                answers = [
                   {"text":text, "answer_start": ans_start}
                   for text, ans_start in zip(ans_text, ans_start_idxs)
                ]
                is_impossible = False
            else:
                answers = []
                is_impossible = True

            # Add question-answer pair to qas
            qas.append({
                "question": question,
                "id": qid,
                "is_impossible": is_impossible,
                "answers":answers,
            })

        paragraphs.append({"qas":  qas, "context": review})

    return paragraphs









````

````python
import json

def convert_to_squad(dfs):
    for split, df in dfs.items():
        subjqa_data = {}
        # Create paragrpah for each id
        groups = (df.groupby("title").apply(create_paragraphs)
        .to_frame(name="paragraphs").reset_index()
        )
        subjqa_data["data"] = groups.to_dict(orient="records")
        # Save the result to disk
        with open(f"electronics-{split}.json", "w+", encoding="utf-8") as f:
            json.dump(subjqa_data, f)
````
